```python
from bs4 import BeautifulSoup as bs
import requests
site = "https://neolook.com"
```

### setMonthlyPageUrl(): 월별 대표 페이지의 주소를 생성하는 함수

```python
def setMonthlyPageUrl(whichYear, whichMonth):
    whichUrl = str(whichYear) + str(whichMonth).zfill(2)
    url = site + "/archives/" + whichUrl 
    return url 
    
```

### getMonthlyPage(): 월별 대표 페이지를 읽어서 반환하는 함수

```python
def getMonthlyPage(whichUrl):
    return requests.get(whichUrl)
 
```

### getMonthlyList(): 월별 대표 페이지에서 개뱔 페이지의 ID를 리스트로 반환하는 함수

```python
def getMonthlyList(whichMonthlyPage):
    soup = bs(whichMonthlyPage.text, "html.parser")
    #soup.prettify()
    #soup.find(class_="mt-4.5 divide-y divide-dim-gray")
    result = soup.find_all("li", class_="py-1")
    #print(result)
    #print(len(result))
    #print(result[0])
    #print(result[0].a["href"])

    volumeList = []
    for volume in result:
        #print(item.a["href"])
        volumeList.append(volume.a["href"])

    return volumeList
```

### saveList(): 해당월의 개별 페이지 리스트를 파일로 저장하는 함수

```python
def saveList(list, fileName, mode='w'):
    with open(fileName, mode) as f:
        for item in list:
            id = item.split("/")[2]
            #print(id)
            f.write(f"{id}\n")
```
